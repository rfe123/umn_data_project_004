```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
games_df = pd.read_csv('data/nfl-big-data-bowl-2024/games.csv')
games_df = games_df.astype({'gameId': 'string',})
games_df.head()
```

```python
# Fill non-finite values with 0
games_df['gameId'] = games_df['gameId'].fillna(0)
# Convert the column to integer
games_df['gameId'] = games_df['gameId'].astype(int)
```

```python
matchups_df = pd.DataFrame({
    'matchup': games_df['homeTeamAbbr'] + '_' + games_df['visitorTeamAbbr'],
    'gameId': games_df['gameId'],
    'week': games_df['week'],
    'home': games_df['homeTeamAbbr'],
    'away': games_df['visitorTeamAbbr'],
    'week': games_df['week']
})
matchups_df.head()
    
```

```python
matchups_df.to_csv('data/nfl-big-data-bowl-2024/matchups.csv', index=False)
```
